# To create a year column by using  date column

```python
#first change datatype of date column frome object to datetime
ipl['date']=pd.to_datetime(ipl['date'])
```

```python
ipl['date']
```

```python
#now extract the year from date
ipl['date'].dt.year
```

```python
#now add new column into data frame 
ipl['year']=ipl['date'].dt.year
```

```python
# count the matches by year 
ipl['year'].value_counts()
```
